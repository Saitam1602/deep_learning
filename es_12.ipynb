```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
z = np.arange(-10,10,.5)
```

```python
def sigmoide(z):
    return 1 / (1 + np.exp(-z))

def sigmoide_derivate(z):
    return sigmoide(z)*(1-sigmoide(z))

def tanh(z):
    return 2*sigmoide(2*z) - 1

def tanh_derivate(z):
    return 1 - (tanh(z))**2

def relu(z):
    return np.maximum(0, z)

def relu_derivate(z):
    return np.heaviside(z,1)
```

```python
plt.figure(figsize=(15,10))
plt.plot(z, tanh(z))
plt.plot(z, tanh_derivate(z))
```

```python
plt.figure(figsize=(15,10))
plt.plot(z, relu(z))
plt.plot(z, relu_derivate(z))
```
