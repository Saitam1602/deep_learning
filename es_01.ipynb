```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ar = np.arange(-10,10,.5)
```

```python
b = np.array([0,4,1,0])
w = np.array([1,1,4,.5])
color_ = ["r", "g", "b", "y"]
```

```python
plt.figure(figsize=(15,10))
for x in range(len(b)):
    plt.plot(ar, 1/(1+(np.e**(-b[x]-w[x]*ar))), color=color_[x], label=f"b = {b[x]}, w = {w[x]}")
    plt.legend()
```
