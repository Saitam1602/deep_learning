# Calcolo del gradiente

### Dati
- formula = 2x^2 -3x +1
- step = 0.01
- x = -7.5
- iterazioni = 1000
- derivata = 4x -3

```python
#
step = 0.01
x = -7.5
iterazioni = 1000
```

```python
def nuovo_passo(x, step):
    return x - (step) * (4 * x -3)
```

```python
lst_x = [x]
for _ in range(iterazioni):
    x = nuovo_passo(x,step)
    lst_x.append(x)
lst_x
```

```python
lst_y = []
for i in lst_x:
    lst_y.append(2 * i**2 -3 * i + 1)
lst_y
```

```python
lst_x = np.array(lst_x)
lst_y = np.array(lst_y)
```

```python
test_x = np.arange(-15,15)
test_y = 2 * test_x**2 -3 * test_x + 1
```

```python
plt.figure(figsize=(15,8))
plt.plot(test_x, test_y)
plt.plot(lst_x,lst_y, '--or')
```
