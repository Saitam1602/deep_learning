```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.concatenate((np.arange(0,5,.3), np.arange(5,9,.3)), axis = None)
y = np.asarray([1 if val >= 5  else 0 for val in x])
x, y
```

```python
b = np.asarray([-8.8, -7, -6])
w = np.asarray([1.82, 3.1, 4])
```

```python
z = b[0] + w[0] * x
z2 = b[1] + w[1] * x
z3 = b[2] + w[2] * x
z,z2,z3
```

```python
y_cap = 1 / (1 + np.e**-z)
y_cap2 = 1 / (1 + np.e**-z2)
y_cap3 = 1 / (1 + np.e**-z3)
```

```python
l = -1/len(x) * (np.dot(y,np.log(y_cap)) + np.dot(1-y, np.log(1-y_cap)))
l2 = -1/len(x) * (np.dot(y,np.log(y_cap2)) + np.dot(1-y, np.log(1-y_cap2)))
l3 = -1/len(x) * (np.dot(y,np.log(y_cap3)) + np.dot(1-y, np.log(1-y_cap3)))
```

```python
l,l2,l3
```

```python
plt.figure(figsize=(15,5))
plt.plot(x, y_cap)
plt.plot(x, y_cap2)
plt.plot(x, y_cap3)
```
